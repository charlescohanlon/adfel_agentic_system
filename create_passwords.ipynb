```python
import secrets
import string
import pandas as pd

from app import ENROLLMENT_FILE

df = pd.read_excel(ENROLLMENT_FILE)


def generate_password(length=16):
    """Generate a strong alphanumeric password."""
    alphabet = string.ascii_letters + string.digits
    return "".join(secrets.choice(alphabet) for _ in range(length))


# Generate unique passwords for each row
df["password"] = [generate_password() for _ in range(len(df))]

# Save to the same Excel file
df.to_excel(ENROLLMENT_FILE, index=False)
```
